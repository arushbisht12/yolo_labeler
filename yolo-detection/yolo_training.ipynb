```python
from ultralytics import YOLO
import os
import torch
import torchvision
```

```python
print(f"Version: {torch.__version__}, GPU: {torch.cuda.is_available()}, NUM_GPU: {torch.cuda.device_count()}")
print(torchvision.__version__)  # should say (version)+cu126 or similar if GPU is detected
print(torch.__version__)

# if device has a GPU, recommend setting up cuda + pytorch version 
# so that pytorch can detect it
device = 'cpu'
if torch.cuda.is_available():
    device = '0'

print(os.getcwd())
```

```python
# To start from new model:
model = YOLO("yolo11n.pt")

# To continue training from a previous model: place in this directory, then:
# model = YOLO("last.pt")

# YOLO automatically handles your 16:9 images by adding gray padding
# to make them square during training. This is the most stable method.
model.train(
    data="data.yaml",
    epochs=400,
    imgsz=256,       # Standard training size (Higher res = better detection)
    batch=16,
    workers=4,       # Increase for better performance but higher RAM use
    name='oakd_lite_model',
    device=device,        # Use '0' for GPU, 'cpu' for CPU
    lr0=0.005,       # new settings applied from here
    resume=False,    # False = new settings, True = exact same settings
    copy_paste=0.2,
    scale=0.5,
    translate=0.1,
    hsv_s=0.7,
    hsv_v=0.4,
    erasing=0.3,
    save_period=50,
    patience=0,
    perspective=0.0005
)


metrics = model.val()

```
